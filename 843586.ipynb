Primeiro carreguei os dados em um DataFrame

```python
import pandas as pd

df = pd.read_csv("diamonds.csv")
df.head()
```

Então limpei eles usando 'dropna' para limpar valores nulos e 'drop_duplicates' para limpar duplicatas

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
import numpy as np

def corr(x, y, hue = None, **kwargs):
    coe = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.text(0.5, 0.5, f'r = {coe:.2f}', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
    ax.set_axis_off()


x = sns.PairGrid(df, hue = "cut")

x.map_diag(sns.histplot)
x.map_lower(sns.scatterplot)
x.map_upper(corr)
```
